```python
import pandas as pd

# Load the dataset (update the path if necessary)
file_path = "superstore.csv"  # Replace with your actual file path
superstore = pd.read_csv(r"C:\Users\saive\OneDrive\Desktop\UNIVERSITY\Data Exploration\Class Notes\superstore.csv")
```

```python
top_cities = superstore['City'].value_counts().head(10).index
```

```python
sns.countplot(data=superstore[superstore['City'].isin(top_cities)], x='City')
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=superstore, x='Sales', y='Shipping.Cost')
plt.title("Scatter Plot: Sales vs Shipping Cost")
plt.xlabel("Sales")
plt.ylabel("Shipping Cost")
plt.show()
```
