Definition:

Analyzing a single variable at a time to understand its distribution, central tendency, and variability.
For Categorical Data:

Look at the frequency of categories.
Use:
Bar plots
Pie charts
Example:

Columns like Category, Segment, Ship.Mode.
For Numerical Data:

Calculate statistics:
Mean, Median, Standard Deviation.
Visualize:
Histogram
Box plot
KDE plot
Example:

Columns like Sales, Discount, Shipping.Cost.
For Date/Time Data:

Analyze trends by extracting components like year, month, or day.
Use line plots or bar plots for time-based trends.
Outlier Detection:

Use box plots to identify extreme values in numerical data.

```python
top_cities = superstore['City'].value_counts().head(10).index
```

```python
sns.countplot(data=superstore[superstore['City'].isin(top_cities)], x='City')
plt.xticks(rotation=30)  
plt.show()
```

```python
sns.scatterplot(data=superstore, x='Sales', y='Shipping.Cost')
plt.show()
```

```python
sns.histplot(data=superstore, x='Sales', bins=30, kde=True)
plt.show()
```

```python
sns.boxplot(data=superstore, x='Sales', y='Shipping.Cost')
plt.show()
```
